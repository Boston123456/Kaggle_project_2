```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("./datasets/train.csv")
```

```python
df["Upc"] = df["Upc"].fillna(0).astype("int64")
```

```python
df.head()
```

```python
fig, ax = plt.subplots(figsize=(12,12))
df["TripType"].dropna().value_counts().plot.bar()
plt.title("TripType")
plt.show()

fig, ax = plt.subplots(figsize=(12,12))
df["Weekday"].dropna().value_counts().plot.bar()
plt.title("Weekday")
plt.show()

fig, ax = plt.subplots(figsize=(12,12))
df["ScanCount"].dropna().value_counts().plot.bar()
plt.title("ScanCount")
plt.ylim(10000)
plt.show()

fig, ax = plt.subplots(figsize=(12,12))
df["DepartmentDescription"].dropna().value_counts().plot.bar()
plt.title("DepartmentDescription")
plt.show()

```
